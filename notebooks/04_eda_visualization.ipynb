```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/processed/master_dataset.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
category_sales = df.groupby("cat_id")["sales"].sum()

category_sales.plot(kind="bar", figsize=(8,5))

plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")

plt.show()
```

```python
store_sales = df.groupby("store_id")["sales"].sum()

store_sales.plot(kind="bar", figsize=(10,5))

plt.title("Sales by Store")
plt.show()
```

```python
price = df["sell_price"].mean()

print("Average Selling Price:", price)
```

```python
top_products = (
    df.groupby("item_id")["sales"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

print(top_products)
```
