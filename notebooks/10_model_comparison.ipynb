```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
results = pd.DataFrame({
    "Model": ["Prophet", "LightGBM"],
    "MAE": [12.5, 8.7]
})
```

```python
results
```

```python
plt.figure(figsize=(6,4))

plt.bar(results["Model"], results["MAE"])

plt.title("Model Comparison (MAE)")
plt.xlabel("Model")
plt.ylabel("Mean Absolute Error")

plt.show()
```

```python
best_model = results.loc[results["MAE"].idxmin(), "Model"]

print("Best Model:", best_model)
```

## Model Comparison

- Prophet provides a strong baseline for time-series forecasting.
- LightGBM leverages engineered features to improve prediction accuracy.
- Based on MAE, the model with the lower error is selected as the final forecasting model.
