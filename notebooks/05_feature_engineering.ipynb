```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/processed/master_dataset.csv")
```

```python
df.head()
df.info()
```

```python
df["date"] = pd.to_datetime(df["date"])
```

```python
df = df.sort_values(["item_id", "store_id", "date"])
```

```python
df["lag_1"] = df.groupby(["item_id", "store_id"])["sales"].shift(1)
```

```python
df["lag_7"] = df.groupby(["item_id", "store_id"])["sales"].shift(7)
```

```python
df["rolling_mean_7"] = (
    df.groupby(["item_id", "store_id"])["sales"]
      .transform(lambda x: x.rolling(7).mean())
)
```

```python
df["rolling_std_7"] = (
    df.groupby(["item_id", "store_id"])["sales"]
      .transform(lambda x: x.rolling(7).std())
)
```

```python
df[
    [
        "sales",
        "lag_1",
        "lag_7",
        "rolling_mean_7",
        "rolling_std_7",
    ]
].head(15)
```

```python
df.fillna(0, inplace=True)
```

```python
df.to_csv(
    "../data/processed/master_dataset_features.csv",
    index=False
)
```
