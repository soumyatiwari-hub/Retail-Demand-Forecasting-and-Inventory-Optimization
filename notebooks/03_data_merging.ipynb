```python
import pandas as pd
import numpy as np
print("Working")
```

```python
sales = pd.read_csv("../data/raw/sales_train_validation.csv")
calendar = pd.read_csv("../data/processed/calendar_cleaned.csv")
prices = pd.read_csv("../data/raw/sell_prices.csv")
```

```python
print("Sales:", sales.shape)
print("Calendar:", calendar.shape)
print("Prices:", prices.shape)
```

```python
sales_long = sales.melt(
    id_vars=[
        "id",
        "item_id",
        "dept_id",
        "cat_id",
        "store_id",
        "state_id"
    ],
    var_name="d",
    value_name="sales"
)
```

```python
sales_long.head()
```

```python
sales_long.shape
```

```python
sales_sample = sales.head(500)
```

```python
sales_long = sales_sample.melt(
    id_vars=[
        "id",
        "item_id",
        "dept_id",
        "cat_id",
        "store_id",
        "state_id"
    ],
    var_name="d",
    value_name="sales"
)
```

```python
merged = sales_long.merge(
    calendar,
    on="d",
    how="left"
)
```

```python
merged.shape
```

```python
merged.head()
```

```python
merged = merged.merge(
    prices,
    on=["store_id", "item_id", "wm_yr_wk"],
    how="left"
)
```

```python
merged.isnull().sum()
```

```python
merged.to_csv(
    "../data/processed/master_dataset.csv",
    index=False
)
```
